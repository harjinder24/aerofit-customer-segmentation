# Business Problem
<a class="anchor" id="business-problem"></a>
Identify the **characteristics of the target audience** and **build a customer profile** for **each type of treadmill** offered by the company, to provide a **better recommendation** of the treadmills to the new customers.

<a class="anchor" id="about-the-dataset"></a>

|:------------------------|:-----------------------|

# Data Visualisation - Overall Dataset

## Age Group Distribution

```python
aero_df.head()
```

```python
bins = [15, 25, 35, 45, 55]
labels = ['15-25', '26-35', '36-45', '46-55' ]

aero_df['AgeGroup'] = pd.cut(aero_df['Age'], bins=bins, labels=labels)
aero_df.head()
```

```python
plt.figure(figsize=(15, 10))
aerofit_palette = ['#243e8d', '#db2a27']
sns.displot(aero_df['AgeGroup'], kde=True, color=aerofit_palette[0])
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Distribution of customers by age group')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**Observation:** Most customers belong to the age group of 21-25. <br> <br>
**Let's look at the gender-wise distribution of age:**

```python
# plotting barplot
sns.countplot(x = aero_df['AgeGroup'], hue='Gender', palette=aerofit_palette, data=aero_df, edgecolor='white', linewidth=1, alpha=0.85)
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Distribution of Customers by Age Group Based on Gender')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**Observation:** 
- Most customers in both genders belong to the age group of 21-25.
- Number of males in every age group is greater then female customers, except 31-35 where number of male and female customers are same.

**Insight:** The most audience for Aerofit exercise bikes belongs to the age group of 21-35.

**Let's look at revenue distribution by age groups and gender**

```python
# plotting barplot
sns.barplot(x=aero_df['AgeGroup'], y=aero_df['Income'], hue='Gender', palette=aerofit_palette, data=aero_df, edgecolor='white', linewidth=1, alpha=0.85)
plt.xlabel('Age Group')
plt.ylabel('Annual Income')
plt.title('Distribution of Customers by Age Group Based on Income and Gender')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**Observation:** 
- The annual income of people belonging to the age group of 41-50 is the highest for both genders.
- However, the number of total customers from 41-50 is the lowest. More exploration should be done on how to get more customers from this age group as they are more likely to spend. 


```python
# Creating Income Groups
inc_bins = [25000, 50000, 75000, 100000, 125000] 
inc_labels = ['\$25,001-\$50,000', '\$50,001-\$75,000', '\$75,001-\$100,000', '\$100,001+']

aero_df['IncomeRange'] = pd.cut(aero_df['Income'], bins=inc_bins, labels=inc_labels)
```

```python
# Income and product purchase
sns.countplot(x=aero_df['IncomeRange'], hue=aero_df['Product'], palette='deep')
plt.xlabel('Income')
plt.ylabel('Purchase Count')
plt.title('Popular Product by Income')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

- People earning between $ \$25000-\$75000 $ are most likely to purchase KP281 followed by KP481.
- People earning more than $ \$75000 $ are most likely to purchase KP781.

**Let's look at Which is the most selling product by age group**

```python
# plotting barplot
sns.countplot(x=aero_df['AgeGroup'], hue='Product', palette='deep', data=aero_df, edgecolor='white', linewidth=1)
plt.xlabel('Age Group')
plt.ylabel('Purchase Count')
plt.title('Popular Product in Each Age Group')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**Observation:**
- The purchase of all three products is highest for age group 21-25.
- For each age-group except 31-35 purchase of KP281, which is the entry-level treadmill is highest.

# Probability Analysis
<a class="anchor" id="probability-analysis"></a>

## Probability of Purchasing A Product Based On Gender

```python
sns.heatmap(pd.crosstab(aero_df['Product'], aero_df['Gender'], normalize=True, margins=True), annot=True, 
           cmap='Blues')
plt.title('Probability of Purchasing A Product Based On Gender')
plt.tight_layout()
plt.show()
```

- Females are least likely to purchase KP781.
- The probaility of buying KP281 is equal among male and female customers.
- The Probability of Purchasing an exercise bike is higher in males as compared to females.

## Probability of Purchasing A Product Based On Marital Status

```python
pms = pd.crosstab(aero_df['Product'], aero_df['MaritalStatus'], normalize=True, margins=True)
pms
```

```python
sns.heatmap(pms, annot=True, cmap='Blues')
plt.title('Probability of Purchasing A Product Based On Marital Status')
plt.tight_layout()
plt.show()
```

- Couples are more likely to purchase a bike as compared to single people. Although there's a mix response in the choice of bikes.

## Probability of Purchasing A Product Based On Age Group

```python
pag = pd.crosstab(aero_df['Product'], aero_df['AgeGroup'], normalize=True, margins=True)
pag
```

```python
sns.heatmap(pag, annot=True, cmap='Blues')
plt.title('Probability of Purchasing A Product Based On Age Group')
plt.tight_layout()
plt.show()
```

- Age Group doesn't seem to affect the product purchase decision.

## Probability of Purchasing A Product Based On Usage

```python
pus = pd.crosstab(aero_df['Product'], aero_df['Usage'], normalize=True, margins=True)
pus
```

```python
sns.heatmap(pus, annot=True, cmap='Blues')
plt.title('Probability of Purchasing A Product Based On Weekly Usage')
plt.tight_layout()
plt.show()
```

- Customers planning to use exercise bikes to use 6 or 7 days per week are least likely to purchase KP281 and KP481.
- Customers planning to use exercise bikes to use 2 or 3 days per week are least likely to purchase KP781.

## Probability of Purchasing A Product Based On Fitness

```python
pf = pd.crosstab(aero_df['Product'], aero_df['Fitness'], normalize=True, margins=True)
pf
```

```python
sns.heatmap(pf, annot=True, cmap='Blues')
plt.title('Probability of Purchasing A Product Based On Fitness')
plt.tight_layout()
plt.show()
```

- Customers who rate themselves 3 are more likely to purchase an exercise bike and prefer KP281 or KP481.

## Probability of Purchasing A Product Based On Miles

```python
aero_df['Miles'].describe()
```

```python
miles_bin = []
for i in range(20, 361, 68):
    miles_bin.append(i)

miles_labels = ['20-88', '89-156', '157-224', '225-292', '293-360']
```

```python
aero_df['MilesGroup'] = pd.cut(aero_df['Miles'], bins=miles_bin, labels=miles_labels)
aero_df.head()
```

```python
pmiles = pd.crosstab(aero_df['Product'], aero_df['MilesGroup'], normalize=True, margins=True)
pmiles
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(pmiles, annot=True, cmap='Blues')
plt.title('Probability of Purchasing A Product Based On Fitness')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

- Customers planning to use exercise bikes for longer miles (157 or higher) prefer KP781.
- Customers planning to use exercise bikes to use 20-156 per week prefer KP281 and KP481.

## Probability of Purchasing A Product Based On Income

```python
pinc = pd.crosstab(aero_df['Product'], aero_df['IncomeRange'], normalize=True, margins=True)
pinc
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(pinc, annot=True, cmap='Blues')
plt.title('Probability of Purchasing A Product Based On Fitness')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

- Customers earning between $ \$25k-\$75k $ are most likely to purchase an exercise bike
